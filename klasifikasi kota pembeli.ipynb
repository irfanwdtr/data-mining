```python
import pandas as pd
```

```python
penjualan = pd.read_csv('penjualan50.csv')
```

```python
data = penjualan.copy()
```

```python
target = 'nama_produk'
```

```python
encode = ['jenis_kelamin']
```

```python
data
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data,dummy], axis=1)
    del data[col]
```

```python
data
```

```python
target_mapper = {'spotify':0, 'youtube':1, 'netflix':2}
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
data['nama_produk'] = data['nama_produk'].apply(target_encode)
# mengubah nilai dalam nama_produk
data
```

data

```python
x = data.drop('nama_produk',axis=1)
y = data['nama_produk']
# memisahkan
# hilangkan kolom 1
```

```python
x
```

```python
y
```

```python
#model naive bayes klasifikasi NBC
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x,y)
```

```python
#menyimpan model NBC
import pickle
pickle.dump(model, open('modelNBC_penjualan.pkl', 'wb'))
```
